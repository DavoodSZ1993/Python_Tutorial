```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data generation
np.random.seed(0)
episodes = np.linspace(0, 100000, 500)  # X-axis: Number of episodes
means = {
    "WeightedSum": np.random.normal(loc=-5, scale=0.5, size=episodes.shape[0]).cumsum(),
    "Chebyshev": np.random.normal(loc=-6, scale=0.5, size=episodes.shape[0]).cumsum(),
    "ReferencePoint": np.random.normal(loc=-7, scale=0.5, size=episodes.shape[0]).cumsum(),
    "DWA": np.random.normal(loc=-8, scale=0.5, size=episodes.shape[0]).cumsum()
}
stds = {
    "WeightedSum": np.random.random(size=episodes.shape[0]) + 0.5,
    "Chebyshev": np.random.random(size=episodes.shape[0]) + 0.5,
    "ReferencePoint": np.random.random(size=episodes.shape[0]) + 0.5,
    "DWA": np.random.random(size=episodes.shape[0]) + 0.5
}
```

```python
colors = {
    "WeightedSum": "blue",
    "Chebyshev": "orange",
    "ReferencePoint": "red",
    "DWA": "purple"
}

plt.figure(figsize=(12, 6))
for key in means:
    plt.plot(episodes, means[key], label=key, color=colors[key])
    plt.fill_between(episodes, means[key] - stds[key], means[key] + stds[key])

plt.title("Mean Reward with Standard Deviation")
plt.xlabel("Episode")
plt.ylabel("Mean Reward")
plt.legend()
plt.grid(True)
plt.show()
```

```python
x = np.arange(0.0, 2, 0.01)
y1 = np.sin(2 * np.pi * x)
y2 = 0.8 * np.sin(4 * np.pi * x)

fig, (ax1, ax2, ax3) = plt.subplots(3, 1, sharex=True, figsize=(6, 6))

ax1.fill_between(x, y1)
ax1.set_title('fill between y1 and 0')

ax2.fill_between(x, y1, 1)
ax2.set_title('fill between y1 and 1')

ax3.fill_between(x, y1, y2)
ax3.set_title('fill between y1 and y2')
ax3.set_xlabel('x')
fig.tight_layout()
```

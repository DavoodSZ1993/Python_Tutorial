```python
%matplotlib inline
import matplotlib.pyplot as plt
```

* `matplotlib.pyplot.imshow(X, cmap=None)`: is used to display data as an  image, i.e., on a 2D regular raster.
* **X**: array-like or PIL (Python Image Library) image.
* **cmap**: The Colormap instance or registered colormap name used to map scalar data to colors. This parameter is ignored if **X** is RGB(A) (Red Green Blue Alpha).

```python
import torch

X = torch.normal(0, 1, (10, 10))

plt.imshow(X, cmap='Reds')
```

* `matplotlib.pyplot.colorbar(mappable=None, ax=None)`: Add a colorbar to a plot.
* **Mappable**: The `matplotlib.cm.ScalarMappable` (i.e., `AxesImage`, `Contourset`, etc) described by this colorbar. This argument is mandatory for the `Figure.colorbar` method but is optional for the `pyplot.colorbar` function.

```python
pcm = plt.imshow(X, cmap='Reds')
plt.colorbar(pcm)
```

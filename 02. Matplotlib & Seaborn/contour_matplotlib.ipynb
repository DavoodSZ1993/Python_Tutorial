## matplotlib.pyplot.contour()
Plots contour lines. More on this on [Matplotlib Documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.contour.html).

```python
%matplotlib inline
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f_2d(x1, x2):
  return x1 **2 + 2 * x2 ** 2
```

```python
  x1, x2 = torch.meshgrid(torch.arange(-5.5,1.0, 0.1),
                          torch.arange(-3.0, 1.0, 0.1))
  plt.contour(x1, x2, f_2d(x1, x2), colors='#1f77b4')
```

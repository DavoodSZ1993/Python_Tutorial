## Color Formats

* RGB or RGBA (red, green, blue, alpha) tuple of float values in closed interval [0, 1]

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = np.sin(np.pi * x)

fig = plt.figure(figsize=(5, 3))
plt.plot(x, y, color=(0.1, 0.2, 0.5))    # (RED, GREEN, BLUE)
```

* Case-insensitive hex RGB or RGBA string. For more on this type of color format, please visit this [website](https://www.hexcolortool.com/#012d83).

```python
x = np.arange(-5.0, 5.0, 0.1)
y = np.sin(np.pi * x)

fig = plt.figure(figsize=(5, 3))
plt.plot(x, y, color='#ff7f0e')
```

* Case-insensitive RGB or RGBA string equivalnet hex shorthand of duplicated characters. e.g., `'#abc'` as `'#aabbcc'`.

```python
x = np.arange(-5.0, 5.0, 0.1)
y = np.sin(np.pi * x)

fig = plt.figure(figsize=(5, 3))
plt.plot(x, y, color='#abc')
```

* String representation of float values in closed interval [0, 1] for grayscale values. `'0'` as black, `'1'` and white, and `'0.8'` as light gray.

```python
x = np.arange(-5.0, 5.0, 0.1)
y = np.sin(np.pi * x)

fig = plt.figure(figsize=(5, 3))
plt.plot(x, y, color='0')
```

* Single character shorthand notation for some basic colors. e.g., `'k'` for black and `'c'` for cyan. Please visit [Matplotlib Documentation](https://matplotlib.org/stable/tutorials/colors/colors.html) for more information.

```python
x = np.arange(-5.0, 5.0, 0.1)
y = np.sin(np.pi * x)

fig = plt.figure(figsize=(5, 3))
plt.plot(x, y, color='c')
```
